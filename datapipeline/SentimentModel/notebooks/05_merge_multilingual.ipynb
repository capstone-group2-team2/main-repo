## Merge

```python
import pandas as pd

ar_train = pd.read_csv("../data/processed/arabic/train.csv")
en_train = pd.read_csv("../data/processed/english/train.csv")

train = pd.concat([ar_train, en_train], ignore_index=True)

train = train.sample(frac=1, random_state=42).reset_index(drop=True)

train.to_csv(
    "../data/finaldataset/train.csv",
    index=False,
    encoding="utf-8-sig"
)
```

```python
ar_valid = pd.read_csv("../data/processed/arabic/validation.csv")
en_valid = pd.read_csv("../data/processed/english/validation.csv")

valid = pd.concat([ar_valid, en_valid], ignore_index=True)

valid = valid.sample(frac=1, random_state=42).reset_index(drop=True)

valid.to_csv(
    "../data/finaldataset/validation.csv",
    index=False,
    encoding="utf-8-sig"
)
```

```python
ar_test = pd.read_csv("../data/processed/arabic/test.csv")
en_test = pd.read_csv("../data/processed/english/test.csv")

test = pd.concat([ar_test, en_test], ignore_index=True)

test = test.sample(frac=1, random_state=42).reset_index(drop=True)

test.to_csv(
    "../data/finaldataset/test.csv",
    index=False,
    encoding="utf-8-sig"
)
```
