```python
import pandas as pd
```

```python
train = pd.read_csv("../data/finaldataset/train.csv")
test = pd.read_csv("../data/finaldataset/test.csv")
valid = pd.read_csv("../data/finaldataset/validation.csv")
```

```python
# عدد الأحرف
train["char_length"] = train["text"].astype(str).apply(len)

# عدد الكلمات
train["word_length"] = train["text"].astype(str).apply(lambda x: len(x.split()))
```

```python
print("=" * 50)
print("Character Length Statistics")
print(train["char_length"].describe())

print("\n")

print("=" * 50)
print("Word Length Statistics")
print(train["word_length"].describe())
```

```python
longest_review = train.loc[train["char_length"].idxmax(), "text"]

print(longest_review)
```

```python
print("Longest Review Characters:", train["char_length"].max())
print("Longest Review Words:", train["word_length"].max())
```
