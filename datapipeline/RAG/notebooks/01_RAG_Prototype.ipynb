# Synova RAG Prototype

## Objective

This notebook builds the first Retrieval-Augmented Generation (RAG) prototype for Synova.

Goals:

- Validate the RAG pipeline.
- Support Arabic and English.
- Test retrieval quality before building the complete knowledge base.
- Use only a small set of technical documents.

```python
from pathlib import Path
```

```python
PROJECT_ROOT = Path.cwd().parent

KB_DIR = PROJECT_ROOT / "knowledge_base"

TECH_EN = KB_DIR / "technical" / "en"
TECH_AR = KB_DIR / "technical" / "ar"

TECH_EN.mkdir(parents=True, exist_ok=True)
TECH_AR.mkdir(parents=True, exist_ok=True)
```

```python
for p in TECH_EN.glob("*.md"):
    print(p.name)

for p in TECH_AR.glob("*.md"):
    print(p.name)
```
