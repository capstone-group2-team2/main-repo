# Synova RAG - Generate Embeddings

## Objective

Build the FAISS vector database and store document metadata.

The workflow includes:

1. Load CSV.
2. Load embeddings.npy.
3. Build FAISS.
4. Save synova.index.
5. Save metadata.pkl.

# 02 Import Libraries

```python
from pathlib import Path
import sys
import pickle

import faiss
import numpy as np
import pandas as pd
```

# 03 Paths

```python
PROJECT_ROOT = Path.cwd().parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))
```

```python
PROCESSED_DIR = PROJECT_ROOT / "processed"
VECTOR_DB_DIR = PROJECT_ROOT / "vector_db"

VECTOR_DB_DIR.mkdir(exist_ok=True)
```

# 04 Load RAG Documents

```python
df = pd.read_csv(
    PROCESSED_DIR / "knowledge_base.csv"
)

print(df.shape)

df.head()
```

# 05 Load Embeddings

```python
embeddings = np.load(
    VECTOR_DB_DIR / "embeddings.npy"
)

print(embeddings.shape)
```

# 06 Build FAISS Index

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatIP(
    dimension
)

index.add(embeddings)

print(
    f"Vectors indexed: {index.ntotal}"
)
```

# 07 Save FAISS Index

```python
faiss.write_index(

    index,

    str(
        VECTOR_DB_DIR / "synova.index"
    )

)

print("FAISS index saved.")
```

# 08 Save Metadata

```python
metadata = df.to_dict(
    orient="records"
)

with open(

    VECTOR_DB_DIR / "metadata.pkl",

    "wb"

) as f:

    pickle.dump(
        metadata,
        f
    )

print("Metadata saved.")
```

# 09 Verify

```python
print(
    f"Documents : {len(metadata)}"
)

print(
    f"Vectors   : {index.ntotal}"
)
```
